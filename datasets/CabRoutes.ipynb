Create cab routes for each of the calculated journey times

```python
import os
import pandas as pd
import numpy as np
import networkx as nx

FOLDER = r'link_times'
ROADS = pd.read_csv('ny_roads.csv',header=None,usecols=[0,1])
GRAPHTYPE = nx.DiGraph()
nodes = np.unique(ROADS.values)
def create_weighted_graph(filenm):
        
    # get the link journey times and join to the
    # road links
    link_times = pd.read_csv(os.path.join(FOLDER,filenm),
                             header=None)
    weighted_links = pd.concat([ROADS,link_times],axis=1)
    weighted_links.columns = ['o','d','jt']
            
    return nx.from_pandas_edgelist(
        weighted_links,source='o',target='d',
        edge_attr='jt',create_using=GRAPHTYPE
    )
```

Now we have a bunch of generators for getting the shortest path

```python
%%time
shortest_gen = dict()
for filenm in os.listdir(FOLDER):
    
    G = create_weighted_graph(filenm)
    
    # create a ref key
    nm,ext = os.path.splitext(filenm)
    key = nm.split('_')[1]
    
    print(key)
    
    # get the shortest paths
    for n in nodes:
        
        paths = nx.single_source_dijkstra_path_length(G,n,weight='jt')
        
        print(paths)
        
        break
        
    break
    
#         shortest_gen[key,] = nx.all_pairs_dijkstra(G,weight='jt')
```
